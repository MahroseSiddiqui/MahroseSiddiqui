## **PLOTTING IN PYTHON**

```python
# libraries 
import pandas as pd 
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# import dataset
khasti = sns.load_dataset('titanic')
khasti
```

```python
khasti.info()
```

```python
khasti.isnull().sum()
```

```python
khasti.survived.value_counts().plot(kind = "bar")
```

```python
sns.countplot(x = "survived", hue = 'sex',data= khasti)
```

```python
plt.scatter(x = "pclass",y = "fare",data= khasti)
```

```python
sns.scatterplot(x = "age",y = "fare",data= khasti,hue="pclass")
# donoo hii numeric hona chahy
```

```python
sns.barplot(x = "sex",y = "fare",data= khasti,hue="pclass")
```

```python
khasti.columns
```

```python
sns.pairplot(khasti)
```

```python
khasti.hist(bins=10 , grid= False,figsize= (20,10))   # ye blue blue koo bins kehty hain
```

```python
sns.boxplot(x = 'sex', y = 'fare',hue='parch' , data= khasti)
```

```python
sns.boxenplot(x = 'sex', y = 'fare',hue='parch' , data= khasti)
```

```python
g =  sns.FacetGrid(khasti,hue = "survived",col = "pclass",margin_titles=True,
                   palette={1:"seagreen",0 : "gray"})
g = g.map(plt.scatter,"fare","age",edgecolor = "w").add_legend()
```

```python
g =  sns.FacetGrid(khasti,hue = "survived",col = "sex",margin_titles=True,
                   palette={1:"purple",0 : "gray"})
g = g.map(plt.scatter,"fare","age",edgecolor = "w").add_legend()
```

```python
sns.stripplot(x = 'sex', y = 'fare',hue='parch' , data= khasti)
```

```python
#heatmap
corr = khasti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("Heat map of titanic ")
```

```python
sns.violinplot(x = 'sex', y = 'fare',hue='parch' , data= khasti)
```
